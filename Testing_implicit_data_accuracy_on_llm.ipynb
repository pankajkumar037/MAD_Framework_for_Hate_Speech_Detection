```python
import pandas as pd
import re
import string
import unicodedata
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Lowercase text
def to_lower(text):
    return text.lower()

# Remove HTML tags
def remove_html(text):
    return BeautifulSoup(text, "lxml").get_text()

# Remove URLs
def remove_urls(text):
    return re.sub(r"http\S+|www\S+|https\S+", '', text)

# Remove mentions and hashtags
def remove_mentions_hashtags(text):
    return re.sub(r'[@#]\w+', '', text)

# Remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

# Remove extra whitespaces
def remove_extra_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()


# Remove accents and special unicode characters
def normalize_unicode(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')

# Remove stopwords (optional depending on task)
def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop_words])
```

```python
def clean_text_pipeline(text):
    text = to_lower(text)
    text = remove_html(text)
    text = remove_urls(text)
    text = remove_mentions_hashtags(text)
    text = remove_punctuation(text)
    text = remove_numbers(text)
    text = normalize_unicode(text)
    text = remove_extra_spaces(text)
    text = remove_stopwords(text)
    return text
```

```python
implicit_df['cleaned_post'] = implicit_df['post'].astype(str).apply(clean_text_pipeline)
```

```python
def build_prompt(records):
    prompt = f"""
    You are a hate sppech detector.
    Classify the following comments into one of the following labels:
     Comments:{records}
    - "implicit_hate"
    - "not_hate"

    Return the result in **JSON format** with this structure for each:
    {{"index":"...",text": "...", "label": "..."}}. Keep the **same index given in input for each  post**.

    **Do this for all posts in input**

  """
    return prompt
```

```python
from pydantic import BaseModel
import json
import pandas as pd
from typing import List


from pydantic import BaseModel

class speech(BaseModel):
  index:int
  text:str
  label:str

# Final list to accumulate parsed predictions
final_records: List[dict] = []

def accuracy_pipeline(records):
    global final_records

    response = client.models.generate_content(
      model="gemini-2.5-flash",
      contents=build_prompt(records),
      config={
          "response_mime_type": "application/json",
          "response_schema": list[speech],
      }
    )

    res=response.text
    parsed_res = json.loads(res)

    final_records.append(parsed_res)
    return parsed_res
```

```python
from tqdm import tqdm


for i in tqdm(range(0,2000, 100)):

    batch_df_k = implicit_df['cleaned_post'].iloc[i:i+100].reset_index()

    #
    records = batch_df_k.rename(columns={"index": "index", "cleaned_post": "text"}).to_dict(orient="records")


    print(f"\nProcessing batch starting at index {i}:")


    import time
    time.sleep(45)

    try:
        accuracy_pipeline(records)
    except Exception as e:
        print(f"❌ Batch {i//25} failed:", e)
        continue
```
